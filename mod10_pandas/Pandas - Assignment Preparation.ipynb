If we want to calculate a given statistic (e.g., mean, median, sum, min, max) for each category in a column (e.g., male/female in the `Sex` column), we can use the `groupby()` method. This method works by:

1. Splitting the data into groups
2. Applying a function to each group independently
3. Combining the results into a data structure

The apply and combine steps are typically done together in Pandas. For example, if we want to calculate the mean for each numerical column in our dataset grouped by `Sex`, we would do the following:

```python
titanic.groupby("Sex").mean()
```

It doesn't make much sense to get the average value of some of our columns, such as `Pclass`. If we are only interested in the average age for each gender, the selection of columns with `[]` is supported as well:

```python
titanic.groupby("Sex")["Age"].mean()
```

Grouping can be done by multiple columns at the same time, just provide the column names as a list to the `groupby()` method. For example, what is the mean ticket fare price for each of the `Sex` and `Pclass` (cabin class) combinations?

```python
titanic.groupby(["Sex", "Pclass"])["Fare"].mean()
```

```python
titanic.groupby("Sex")[["Age", "Fare"]].mean()
```

We can apply functions over rows of data using the `apply()` method.

Syntax: `DataFrame.apply(function)`

For example:

```python
# create a function
def sex_to_num(x):
    if x == 'male':
        return 1
    if x == 'female':
        return 0

# apply the function to the Sex column and create a new column
titanic['Sex_N'] = titanic["Sex"].apply(sex_to_num)

titanic[["Sex","Sex_N"]].head()
```

```python
titanic
```

If we want to get the index location of a value in a column, we can use the `idxmax()` method. For example, if we wanted to get the passenger with the longest name on the Titanic, we could find its index location by:

```python
titanic["Name"].str.len().idxmax()
```

By using the `str.len()` method, the function is applied to each of the names individually. With this information, we are able to utilize the `loc` operator to determine what the longest name is:

```python
titanic.loc[titanic["Name"].str.len().idxmax(), "Name"]
```

We can filter specific rows from a `DataFrame` using our `[]` notation. For example, if we wanted to get all entries where `Age` is greater than 35, we could write:

```python
titanic[titanic["Age"] > 35].head()
```

```python
titanic["Age"] > 35
```

To select rows based on a conditional expression, we use a condition inside the `[]`. The condition inside the `[]` `titanic['Age'] > 35` checks for which rows the `Age` column has a value larger than 35. 

The output of the conditional expression (>, ==, !=, <, <=, >=, ...) is actually a Pandas `Series` of boolean values (`True` or `False`) with the same number of rows as the original `DataFrame`. Such a `Series` of boolean values can be used to filter the `DataFrame` by putting it in between the `[]`. Only rows for which the value is `True` will be selected.

If we want to combine multiple conditional statements, each condition must be surrounded by parentheses. Moreover, we cannot use `or` or `and`, but we need to use the operators `|` and `&`, respectively:

```python
titanic[(titanic["Age"] > 35) | (titanic["Fare"] < 600)].head()
```

To select multiple columns, use a list of column names within the `[]`. The inner square brackets define a Python list with column names, whereas the outer brackets are used to select the data from a Pandas `DataFrame`:

```python
titanic[["Age","Sex"]].head()
```

Additionally, we can use the `concat()` function, which performs concatenation operations of multiple tables along one of the axes (row-wise or column-wise). By default, concatenation is along axis 0, so the resulting table combines the rows of the input tables. 

For example, if we wanted to combine our two previous datasets together along the rows:

We can also merge datasets together using the `merge()` function. This differs from the `append()` method and the `concat()` function as the `merge()` function need a common column, which is used as a key to combine the information.

Syntax: `pd.merge(DataFrame1, DataFrame2, how, on)`

Parameters:

- `how`: specifies how to merge; default: `"inner"`
    - Other options include  `"left"`, `"right"`, and `"outer"`
- `on`: specifies on what level to do the merging

For example:

```python
data1 = {"name": ["Sally", "Mary", "John"],
         "age": [50, 40, 30]}

data2 = {"name": ["Sally", "Peter", "Micky"],
         "age": [77, 44, 22]}

df1 = pd.DataFrame(data1)
df2 = pd.DataFrame(data2)

newdf = pd.merge(df1, df2, how='left', on='name')

newdf.head()
```
